## Matplotlib example Workflow

```python
# 0 import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#. 1 prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2 Setup plot
fig, ax = plt.subplots(figsize=(4,4))

# 3 Plot the data

ax.plot(x,y)


#4.- Customize plot
ax.set(title='Simple plot',
        xlabel='x-axis',
        ylabel='y-axis')

fig.savefig('images/sample-plot.png')
```

```python
nut_butter_prices = {
    "Almond butter" : 10,
    "Peanut butter" : 8,
    "Cashew butter" : 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values(), color=['red', 'black', 'yellow'])
ax.set(title='Brian store', ylabel='Price', xlabel="Type of butter");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

 ### HOW TO MAKE MULTIPLE PLOT IN SAME FIGURE
    

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
# ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values(), color=['red', 'black', 'yellow'])
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
#plot to each different index

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(100), np.random.random(100))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### wich one should you use pyplot or matplotlib OO method

```python
 over_50 = heart_diasese[heart_diasese['age'] > 50]
```

```python
over_50
```

```python
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
#ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax.set(title='Heart Disease and Cholesterol Levels', xlabel='Age', ylabel='Cholesterol')
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
plt.style.use('seaborn-whitegrid')
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True)

scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')
ax0.set(title='Heart Disease and cholesterol levels', ylabel='Cholesterol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')


scatter_thalach = ax1.scatter(x=over_50['thalach'], y=over_50['chol'], c=over_50['target'],cmap='winter')
ax1.set(title='Cholesterol vs Heart rate', xlabel='Age', ylabel='Cholesterol')
ax1.legend(*scatter_thalach.legend_elements(), title='Target')
ax1.axhline(y=over_50['chol'].mean(), linestyle='--')
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
